```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
driver = webdriver.Edge() 
file = open("data.csv", "w", encoding="utf-8")
file.close()
car_list = []

for i in range(1,20):
    URL="https://www.cars.com/shopping/results/?dealer_id=&keyword=&list_price_max=&list_price_min=&makes[]=mazda&maximum_distance=20&mileage_max=&monthly_payment=&page="+str(i)+"&page_size=100&sort=best_match_desc&stock_type=all&year_max=&year_min=&zip="
    driver.get(URL)
    time.sleep(8)
    
    for i in range(3,95):

        stock_path = f"/html/body/section/div[2]/div[4]/div[2]/div[1]/div[{i}]/div/div[2]/p" 
        marka_path = f"/html/body/section/div[2]/div[4]/div[2]/div[1]/div[{i}]/div/div[2]/a/h2" 
        mıl_path = f"/html/body/section/div[2]/div[4]/div[2]/div[1]/div[{i}]/div/div[2]/div[2]"
        prıce_path = f"/html/body/section/div[2]/div[4]/div[2]/div[1]/div[{i}]/div/div[2]/div[3]/span"

        try:       
            stock_name = driver.find_element(By.XPATH, stock_path).text
            marka_name = driver.find_element(By.XPATH, marka_path).text
            mıl_name = driver.find_element(By.XPATH, mıl_path).text
            prıce_name = driver.find_element(By.XPATH, prıce_path).text.strip()
        
    
            driver.execute_script("window.scrollTo(0,document.body.scrollHeight-1500)")
            time.sleep(0.25)
        
            car_list.append([stock_name,marka_name,mıl_name,prıce_name])
            print(stock_name,marka_name,mıl_name,prıce_name)                    

        except Exception as e:
            print(f"{i}. veri alınırken hata oluştu: {e}")

df = pd.DataFrame(car_list,columns=["kullanım durumu","model","mil","fiyat"])
df.to_csv("data.csv",index=False,encoding="utf-8")

driver.quit()
```
